```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model, model_selection
```

```python
from sklearn.datasets import load_linnerud
linnerud_data = load_linnerud()
```

```python
linnerud_df = pd.DataFrame(
    linnerud_data.data, columns=linnerud_data.feature_names)
linnerud_df.head()
```

```python
linnerud_df_tgt = pd.DataFrame(linnerud_data.target, columns=linnerud_data.target_names)
linnerud_df_tgt.head()
```

```python
tmp = linnerud_df.join(linnerud_df_tgt)
tmp.head()
display(pd.plotting.scatter_matrix(tmp))
```

```python
X = linnerud_df["Situps"].to_numpy().reshape(-1, 1)
y = linnerud_df_tgt["Waist"].to_numpy().reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
